```python
# Tratamiento de datos
# -----------------------------------------------------------------------
import numpy as np
import pandas as pd
import random 

# Gráficos
# ------------------------------------------------------------------------------
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams["figure.figsize"] = (10,8) # para establecer el tamaño de todas las figuras del jupyter al mismo tamaño. 

```

```python
df = pd.read_csv('data/sin_outliers_data.csv', index_col=0)
```

### 1. Pregunta teórica: ¿Qué diferencia existe entre la covarianza y la correlación?

La correlación es si una variable esta relacionada con otra. La covarianza es la dirección de la relación de las variables. 


### 2. Calculad la covarianza y la correlación de nuestro dataset. No nos vale solo con que la calculéis, debéis hacer una interpretación de los resultados obtenidos. 


```python
df.corr() #Calculamos la correlación
```

Nuestras variables no presentan mucha correlación, solo el precio del seguro médico ('charges') con la edad ('age')

```python
df.cov()
```

```python
sns.pairplot(df)
```

No hay una covarianza clara entre las varaibles. Las variables 'charges' y 'age' si presentan una covarianza positiva, pero no muy fuerte.

 ### 3. Calculad solo la correlación de Pearson.

```python
#En el punto 2 hemos calculado la correlación de Pearson, pasamos a hacer el heatmap.

mask = np.triu(np.ones_like(df.corr(), dtype= bool))

```

```python
sns.heatmap(df.corr(), cmap = 'YlGnBu', mask  = mask, annot = True);
```

Ninguna de nuestras variables guarda una relación significativa.
